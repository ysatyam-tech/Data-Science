```python
# Define a generator function to yield the powers of 2 up to a given exponent
def powers_of_2_generator(exponent):
    power = 0  # Initialize the power to 0
    while power <= exponent:  # Continue generating powers of 2 until the power reaches the specified exponent
        yield 2 ** power  # Yield the next power of 2
        power += 1  # Increment the power for the next iteration

# Specify the exponent for the powers of 2
exponent = 5

# Generate and print the powers of 2 up to the specified exponent
for result in powers_of_2_generator(exponent):
    print(result)  # Print each power of 2 generated by the generator function
```
